```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv("TravelInsurancePrediction.csv")
df
```

```python
print("Com esse segundo dataset o que mais me chamou atenção foi pensar sobre correlação, sobre alguma coisa implicava em ter mais ou menos chance de alguém comprar o seguro de viagem, pensando nisso, a primeira pergunta é:")
print()
print("O que mais influencia alguém a comprar seguro de viagem?")

df1=df[["AnnualIncome","ChronicDiseases","FrequentFlyer","EverTravelledAbroad","TravelInsurance"]]

df1.corr().style.background_gradient(cmap="Blues")
```

```python
print("Usando a função do pandas de correlação obtivemos esse resultado;")
print("Dadas as correlações, a frequência de viagens e a distância das viagens tem grande influência por levarem a maior exposição a risco de acidentes em viagens; A correlação com a renda anual também é alta por conta de que é necessário ter uma boa renda para conseguir efetuar várias viagens.")
print()
print("A Segunda pergunta, seguindo na mesma linha de pensamento, é sobre: por que algumas coisas não tem quase nenhuma correlação com contratar um seguro de viagem?")
```

```python
df2=df[["EmploymentType","ChronicDiseases","Age","GraduateOr0t","FamilyMembers","AnnualIncome","TravelInsurance"]]

df2.corr().style.background_gradient(cmap="Reds")
```
